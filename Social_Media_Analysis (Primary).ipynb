```python
df = df.rename(columns={
    'How many followers do you have on Instagram? (In case of multiple accounts, please mention the one with the maximum)':'Instagram Followers',
    'How many posts do you have on Instagram?':'Instagram Posts'
})
df.columns
```

```python
# Since maximum of numerical columns are Object, Below is a function to convert object to numeric
def convert_to_num(x):
    x = x.replace(',','')
    return pd.to_numeric(x)
```

```python
df_c = df[df['Current Status']=='Working professional']
df_c = df[['Total Week Usage','Total Weekend Usage']].applymap(convert_to_num)
```

```python
import matplotlib.pyplot as plt

plt.hist([df_c['Total Week Usage'], df_c['Total Weekend Usage']],label=['Total Week Usage','Total Weekend Usage'])
plt.legend(loc='upper right')
plt.show()
```

```python
plt.boxplot(df_c, showmeans=True)
```

```python
df.drop(df.index[1], inplace=True)
```

```python
plt.figure(figsize=(50,20))
plt.title("City and Age usage data compilation")
sns.scatterplot(x = df['City'],y = df['Age'])
```

Age Group Analysis

```python
ax = df['Age'].value_counts().sort_index().plot(kind='bar', title='Count of age group spend on instagram', figsize=(10,5))
ax.set_xlabel('Age Analysis')
plt.show()
```

```python
df.dropna(inplace=True)
```

```python
ax = df['City'].value_counts().sort_index().plot(kind='bar', title='Count of City  group spend on instagram', figsize=(10,5))
ax.set_xlabel('City Analysis')
plt.show()
```
